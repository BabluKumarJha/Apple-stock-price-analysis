# Apple stock price

```python
# import neccesery library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.dates as mdates

%matplotlib inline
```

```python
# open file in pandas
df = pd.read_csv("apple_stock_price.csv")
print(df)
```

```python
# parse_date in my case already date is in the form of date.
# parse date change['column'] data into date.

# make date column as index. so we will call `index_col` function.

df = pd.read_csv("apple_stock_price.csv", parse_dates = ['Date'], index_col= 'Date')
print(df)
```

```python
#plt.figure(figsize=(10,10))
plt.plot(df.index, 'Open', data = df)
plt.title("Stock Opening Price", fontsize = 30)

plt.show()
```

```python
# show first August 2024 to October 1st 2024.

df.sort_index().loc["2024-08-01":"2024-10-01"].head()
```

```python
# jan month averge stock price.
df.loc['2025-01'].head()
```

```python
# Check monthly Closing stock price of Apple stock price.

df.Close.resample("ME").mean() # here ME represent month.
```

# Visualize daily opening and closing price mean and compare.

```python
plt.figure(figsize= (12,8))
df.Close.plot(title = 'Daily closing price')
```

```python
plt.figure(figsize= (12,8))
df.Open.plot(title = 'Daily Opening Price')
```

```python
# Visualization with mini customization.

fig, axes = plt.subplots(1,2, figsize = (14,7))
df.Open.plot(ax = axes[0])
axes[0].set_title("Opening Pirce", fontsize = 20)
axes[0].grid()

df.Close.plot(ax = axes[1])
plt.suptitle('Opening price vs Closing price', fontsize = 25)
axes[1].set_title('Closing Price', fontsize = 20)
axes[1].grid()

plt.tight_layout()
plt.show()
```

```python
# compare in single plot.
plt.figure(figsize = (12,7))
plt.plot(df.index, df.Open, color = 'r', label = "Opening Price")
plt.plot(df.index, df.Close, color = 'b', label = "Closing Price")

plt.title("Opening price Vs Closing Price", fontsize = 30, color = 'b')
plt.xlabel('Date', fontsize = 20)

plt.legend()
plt.show()
```

# Lets compare Opening and closing average price.

```python
fig, axes = plt.subplots(1,2, figsize = (14,7))

df.Open.resample("YE").mean().plot(ax = axes[0]) # opening yearly average.
df.Close.resample('YE').mean().plot(ax = axes[1]) # closing yearly average.

# all customization for opening price.
axes[0].set_title("Yearly Opening price average", fontsize = 15)
axes[0].set_xlabel('Year', fontsize = 15)
axes[0].set_ylabel("Average Price", fontsize = 15)

# for Closing price.
axes[1].set_title("Yearly, Closing price average", fontsize = 15)
axes[1].set_xlabel('Year', fontsize = 15)
axes[1].set_ylabel("Average Price", fontsize = 15)

plt.show()
```

```python
plt.figure(figsize = (14,7))

plt.plot(df.Open.resample('YE').mean(), color = 'b', label = 'Opening Price')
plt.plot(df.Close.resample('YE').mean(), color = 'r', label = 'Closing Price')

plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.plot(df.Open.resample('W').mean(), color = 'g', label = "Opening Price") # Opening price.
plt.plot(df.Close.resample('W').mean(), color = 'r', label = "Closing Price") # Closing Price.

plt.title('Weekly stock price Opening and closing.', fontsize = 25)

plt.grid() # add gridlines
plt.legend() # show labels
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (14,7))

df.Open.resample('W').mean().plot(ax= axes[0]) # opening price average.
df.Close.resample('W').mean().plot(ax = axes[1]) # Closing price average.

# all customization for weekly stock price.
# all customization for opening price.
axes[0].set_title("Weekly Opening price average", fontsize = 15)
axes[0].set_xlabel('Year', fontsize = 15)
axes[0].set_ylabel("Average Price", fontsize = 15)

# for Closing price.
axes[1].set_title("Weekly, Closing price average", fontsize = 15)
axes[1].set_xlabel('Year', fontsize = 15)
axes[1].set_ylabel("Average Price", fontsize = 15)

plt.show()
```

# Lets check quartly chart and plot.

```python
plt.figure(figsize = (14,7))
df.Close.resample('QE').mean().plot(title= 'Quarterly Closing price')
```

```python
# compare with opening price.
```

```python
fig, axes = plt.subplots(1,2, figsize = (14,7))
df.Open.resample('QE').mean().plot(ax = axes[0])
df.Close.resample('QE').mean().plot(ax = axes[1])

# all customization for Quartly stock price.
# all customization for opening price.
axes[0].set_title("Quarterly Opening price average", fontsize = 15)
axes[0].set_xlabel('Year', fontsize = 15)
axes[0].set_ylabel("Average Price", fontsize = 15)

# for Closing price.
axes[1].set_title("Quarterly, Closing price average", fontsize = 15)
axes[1].set_xlabel('Year', fontsize = 15)
axes[1].set_ylabel("Average Price", fontsize = 15)
plt.show()
```

```python
import matplotlib.dates as mdates

plt.figure(figsize = (12,7))
plt.plot(df.Open.resample('QE').mean(), color = 'y', label = 'Opening price')
plt.plot(df.Close.resample('QE').mean(), color = 'r', label = 'Closing Price')

# Set x-axis ticks to quarterly
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Show every 3rd month
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))  # Format as 'Jan 2024'

plt.title("Quartely stock price comparison.", fontsize = 25) 
plt.xlabel('Date/Quarter')
plt.ylabel('Price')


plt.grid()
plt.legend()
plt.show()
```

# Volume Vs Date

```python
# Convert Volume to numeric if needed
df['Volume'] = df['Volume'].replace(',', '', regex=True).astype(float)  

# Sort by Volume in descending order. latter we will apply filter according to need.
df_sorted = df.sort_values(by='Volume', ascending=False)

# Plot heatmap
# heatmap for top 100 volume date.
plt.figure(figsize=(10, 7))
sns.heatmap(df_sorted[['Volume']].head(100))  # Use by default color.

plt.title("Heatmap of Trading Volume (Sorted in Descending Order)")
plt.show()
```
